```python
import numpy as np
import pickle
from numpy.linalg import norm
from matplotlib import pyplot as plt
```

```python

with open("high-dt-e-migration.pickle", "rb") as file:
    data = pickle.load(file)

with open("amplitudes.pickle", "rb") as file:
    amps = pickle.load(file)

data.to_elements()
```

amps[0]

```python
dPs = amps[1]
amps = amps[0]
```

```python
fig = plt.figure()
plt.scatter(np.real(amps[:,0]), np.imag(amps[:,0]), s=0.01, alpha=0.1)
plt.xlabel('Re(a_tilde)')
plt.ylabel('Im(a_tilde)')
plt.tight_layout()
plt.savefig('vZLK-DT-225-amplitudes.png', dpi=150)
```

```python
e = np.array([data.vectordata.e1x, data.vectordata.e1y, data.vectordata.e1z])
a = data.elementdata.a1
t = data.vectordata.time
e_mags = np.sum(np.abs(e)**2,axis=0)**(1/2)
```

```python
len(a)
```

```python
max(dPs)
```

```python
plt.scatter(a * (1-e_mags), np.log10(dPs), s=1)
plt.xlim([0.01, 0.03])
```

```python
plt.plot(np.log10(t / (1e6 * 365.25)), np.log10(1 - e_mags))
plt.ylabel('log(1-e)')
plt.xlabel('log(t)')
plt.tight_layout()
plt.savefig('vZLK-DT-225-e.png', dpi=150)
```

```python
plt.plot(np.log10(t / (1e6 * 365.25)), np.log10(data.elementdata.a1))
plt.ylabel('log(a)')
plt.xlabel('log(t)')
plt.tight_layout()
plt.savefig('vZLK-DT-225-a.png', dpi=150)
```
